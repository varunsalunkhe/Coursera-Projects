```python
x1=data.loc[:,data.columns!='class']
x=pd.get_dummies(x1)
x.head()
```

```python
y=data['class'].map({'p':0,'e':1})
y.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier , GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
```

```python
train_x, test_x, train_y, test_y= train_test_split(x,y,test_size=0.4, random_state=10)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
model1 = LogisticRegression(penalty='l1', solver='liblinear')
model1.fit(train_x, train_y)
```

```python
p1=model1.predict(test_x)
accuracy_score(p1, test_y)
```

```python
cm1 = confusion_matrix(p1, test_y)
sns.heatmap(cm1, annot= True)
```

```python
model2 = RandomForestClassifier(bootstrap=False, max_depth=3, n_estimators=300)
model2.fit(train_x, train_y)
p2=model2.predict(test_x)
print(accuracy_score(p2, test_y))
cm2 = confusion_matrix(p2, test_y)
sns.heatmap(cm2, annot= True)
```

```python
model3 = AdaBoostClassifier(algorithm='SAMME',
                   base_estimator=DecisionTreeClassifier(max_depth=1),
                   learning_rate=1, n_estimators=100)
model3.fit(train_x, train_y)
p3=model3.predict(test_x)
print(accuracy_score(p3, test_y))
cm3 = confusion_matrix(p3, test_y)
sns.heatmap(cm3, annot= True);
```

```python
model4 = GradientBoostingClassifier(max_depth=2, n_estimators=300)
model4.fit(train_x, train_y)
p4=model4.predict(test_x)
print(accuracy_score(p4, test_y))
cm4= confusion_matrix(p4, test_y)
sns.heatmap(cm4, annot= True);
```

```python
model5 = SVC(C=1, gamma=0.05)
model5.fit(train_x, train_y)
p5=model5.predict(test_x)
print(accuracy_score(p5, test_y))
cm5= confusion_matrix(p5, test_y)
sns.heatmap(cm5, annot= True);
```

```python
model6 = KNeighborsClassifier()
model6.fit(train_x, train_y)
p6=model6.predict(test_x)
print(accuracy_score(p6, test_y))
cm6= confusion_matrix(p6, test_y)
sns.heatmap(cm6, annot= True);
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=100, criterion='gini')
model.fit(x,y)
```
